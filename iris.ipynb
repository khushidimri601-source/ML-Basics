```python
import csv
import random
def LoadSet(filename,split,trainingSet=[],testingSet=[]):
  with open(filename,'r')as csv_file:
    lines=csv.reader(csv_file)
    next(lines)
    dataset=list(lines)
    for x in range(len(dataset)-1):
      for y in range(4):
        dataset[x][y]=float(dataset[x][y])
      if random.random() <split :
        trainingSet.append(dataset[x])
      else:
        testingSet.append(dataset[x])
trainingSet=[]
testingSet=[]
LoadSet(r'C:\Users\hp\Desktop\ML\iris.csv',0.66,trainingSet,testingSet)
print('Train:'+repr(len(trainingSet)))
print('Test:'+repr(len(testingSet)))
```

```python
import math
def eucliden(inst1,inst2,length):
  distance=0
  for x in range(length):
    distance+= pow((inst1[x]-inst2[x]),2)
  return math.sqrt(distance)
d1=[2,2,2,'a']
d2=[4,4,4,'b']
distance=eucliden(d1,d2,3)
print('Distance',repr(distance))
import operator
def GetNgbr(trainingSet,testInstance,k):
  distance=[]
  length=len(testInstance)-1
  for x in range(len(trainingSet)):
    dist=eucliden(testInstance,trainingSet[x],length)
    distance.append((trainingSet[x],dist))
  distance.sort(key=operator.itemgetter(1))
  ngbr=[]
  for x in range(k):
    ngbr.append(distance[x][0])
  return ngbr
```

```python
import operator
def GetNgbr(trainingSet,testInstance,k):
  distance=[]
  length=len(testInstance)-1
  for x in range(len(trainingSet)):
    dist=eucliden(testInstance,trainingSet[x],length)
    distance.append((trainingSet[x],dist))
  distance.sort(key=operator.itemgetter(1))
  ngbr=[]
  for x in range(k):
    ngbr.append(distance[x][0])
  return ngbr
trainSet=[[2,2,2,'a'],[4,4,4,'b']]
testInstance=[5,5,5]
k=1
ngbr=GetNgbr(trainSet,testInstance,1)
print(ngbr)
```

```python
import operator
def GetRes(ngbr):
  ClassVotes={}
  for x in range(len(ngbr)):
    res=ngbr[x][-1]
    if res in ClassVotes:
      ClassVotes[res]+=1
    else:
      ClassVotes[res]=1
  SortedVotes=sorted(ClassVotes.items(),key=operator.itemgetter(1),reverse=True)
  return SortedVotes[0][0]
```

```python
ngbr=[[1,1,1,'a'],[2,2,2,'b'],[3,3,3,'c']]
print(GetRes(ngbr))
```

```python
def getAccuracy(testSet,predictions):
  correct=0
  for x in range(len(testSet)):
    if testSet[x][-1]== predictions[x]:
      correct+=1
  return (correct/float(len(testSet)))*100
testSet=[[1,1,'a'],[2,2,'a'],[3,3,'c']]
prediction=['a','a','a']
accuracy=getAccuracy(testSet,prediction)
print(accuracy)
```

```python
def main():
  trainingSet=[]
  testingSet=[]
  split=0.67
  LoadSet(r'C:\Users\hp\Desktop\ML\iris.csv',split,trainingSet,testingSet)
  print('Training Set'+ repr(len(trainingSet)))
  print('Testing Set'+ repr(len(testingSet)))
  prediction=[]
  k=3
  for x in range(len(testingSet)):
    ngbr=GetNgbr(trainingSet,testingSet[x],k)
    result=GetRes(ngbr)
    prediction.append(result)
    print('predicted='+repr(result)+'actual result'+repr(testingSet[x][-1]))
  accuracy=getAccuracy(testingSet,prediction)
  print(repr(accuracy))
main()

```
